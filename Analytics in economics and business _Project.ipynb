```python
from collections import Counter
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
from matplotlib import pyplot
from numpy import where
from sklearn import decomposition
```

```python
pca = decomposition.PCA()

pca.n_components = 2
pca_data_1 = pca.fit_transform(X_train)

counter = Counter(y_train)
for label, _ in counter.items():
	row_ix = where(y_train == label)[0]
	pyplot.scatter(pca_data_1[row_ix, 0], pca_data_1[row_ix, 1], label=str(label))
pyplot.legend()
pyplot.show()
```

SMOTE algorithm [Nitesh Chawla, et al. (2002)](https://www.jair.org/index.php/jair/article/view/10302/24590) is ran to artificially generate new examples of the minority class (class 1) using the K-nearest neighbors ($K = 5$ by default):

- Step 1: For each $x \in A$ (the minority class), the k-nearest neighbors of x are obtained by calculating the Euclidean distance between x and every other sample in set A.
- Step 2: For each $x \in A$, n examples are randomly selected from its k-nearest neighbors, and they construct the set $A_1$ .
- Step 3: For each example $x_k \in A_1$ (k=1, 2, 3…N), the following formula is used to generate a new example:
$x' = x + rand(0, 1) * \mid x - x_k \mid$ in which rand(0, 1) represents the random number between 0 and 1. (in other words: computes a line between the minority data points and any of its neighbors and places a synthetic point somewhere the middle.)

Furthermore, the majority class (class 0) is also under-sampled, leading to a  balanced dataset.




```python
# Oversample with SMOTE and random undersample for imbalanced dataset

# define pipeline
over = SMOTE(sampling_strategy=0.1)
under = RandomUnderSampler(sampling_strategy=1)
steps = [('o', over), ('u', under)]
#steps = [ ('u', under)]
pipeline = Pipeline(steps=steps)
# transform the dataset
X_train_bal, y_train_bal = pipeline.fit_resample(X_train, y_train)
```

```python
pca.n_components = 2
pca_data_bal = pca.fit_transform(X_train_bal)

counter = Counter(y_train_bal)
for label, _ in counter.items():
	row_ix = where(y_train_bal == label)[0]
	pyplot.scatter(pca_data_bal[row_ix, 0], pca_data_bal[row_ix, 1], label=str(label))
pyplot.legend()
pyplot.show()
```

# Keras Neural Network

```python
import tensorflow as tf
from tensorflow import keras
import keras.backend as K
from keras.callbacks import Callback,ModelCheckpoint
import os
import tempfile

METRICS = [

      keras.metrics.BinaryAccuracy(name='accuracy'),
      keras.metrics.Precision(name='precision'),
      keras.metrics.Recall(name='recall'),
]

```

```python
def make_model(metrics=METRICS):
  
  
  model = keras.Sequential([
      keras.layers.Dense(50, activation='relu', input_shape=(X_train_bal.shape[-1],)),
      keras.layers.Dropout(0.8),

      keras.layers.Dense(5, activation='relu'),
      keras.layers.Dropout(0.8),

      keras.layers.Dense(1, activation='sigmoid'),
     
  ])

  model.compile(
      optimizer=keras.optimizers.Adam(lr=1e-3),
      loss=keras.losses.BinaryCrossentropy(),
      #metrics=[get_f1])
      metrics=METRICS)


  return model
```

```python
EPOCHS = 100
BATCH_SIZE = 50

model = make_model()
model.summary()
```

```python
tf.keras.utils.plot_model(model,  show_shapes=True)
```

This model consists of two hidden layers with relu activatio function $[g(z) = max(0,z)]$, an input layer and an output layer with a sigmoid  activation fucntion. Dropout layers are added in between each pair of dense layers for regularisation. The Dropout layer takes and argument “rate”, which specifies the proportion of neurons in the preceding dense layer that should take a value of zero. In this model, the rate is set to 0.8, which means 80% of the neurons in the hidden layers are given a weight of 0.

 Run the model

```python
model = make_model()

history = model.fit(
    X_train_bal,
    y_train_bal,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    validation_data=(X_val, y_val), 
    #callbacks=[early_stopping],
    verbose=1)
```

Function to plot the history

```python
def plot_metrics(history):
  metrics = ['loss', 'accuracy', 'precision', 'recall']
  for n, metric in enumerate(metrics):
    name = metric.replace("_"," ").capitalize()
    plt.subplot(2,2,n+1)
    plt.plot(history.epoch, history.history[metric],  label='Train')
    plt.plot(history.epoch, history.history['val_'+metric],
              linestyle="--", label='Val')
    plt.xlabel('Epoch')
    plt.ylabel(name)
    if metric == 'loss':
      plt.ylim([0, plt.ylim()[1]])
    elif metric == 'auc':
      plt.ylim([0.8,1])
    else:
      plt.ylim([0,1])

    plt.legend()
```

```python
plt.rcParams["figure.figsize"] = (10,10)
plot_metrics(history)
```

```python
test_predictions = model.predict(X_test, batch_size=BATCH_SIZE)
```

```python
def plot_cm(labels, predictions, p=0.5):
  cm = confusion_matrix(labels, predictions > p)
  plt.figure(figsize=(5,5))
  sns.heatmap(cm, annot=True, fmt="d")
  plt.title('Confusion matrix @{:.2f}'.format(p))
  plt.ylabel('Actual label')
  plt.xlabel('Predicted label')
```

```python
results = model.evaluate(X_test, y_test,
                                  batch_size=BATCH_SIZE, verbose=0)
for name, value in zip(model.metrics_names, results):
  print(name, ': ', value)
print()

plot_cm(y_test, test_predictions)
```

#Functional API

The Keras functional API is a way to create models that are more flexible than the tf.keras.Sequential API. The functional API can handle models with non-linear topology, shared layers, and even multiple inputs or outputs.

```python
from keras.layers.merge import concatenate
from tensorflow.keras import layers


def make_model_2( ):
    
    input = tf.keras.Input(shape=X_train_bal.shape[-1])


    ## Dense layer
    layer1 = tf.keras.layers.Dense(units= 95, activation='relu')(input)
    drop1 = tf.keras.layers.Dropout(0.8)(layer1)

    layer1a = tf.keras.layers.Dense(units= 25, activation='relu')(drop1)
    drop1a = tf.keras.layers.Dropout(0.8)(layer1a)

    layer2 = tf.keras.layers.Dense(units= 50, activation='relu')(input)
    drop2 = tf.keras.layers.Dropout(0.8)(layer2)

    layer2a = tf.keras.layers.Dense(units= 5, activation='relu')(drop2)
    drop2a = tf.keras.layers.Dropout(0.8)(layer2a)


    layer3 = tf.keras.layers.Dense(units= 5, activation='relu')(input)
    drop3 = tf.keras.layers.Dropout(0.8)(layer3)

    layer3a = tf.keras.layers.Dense(units= 5, activation='relu')(drop3)
    drop3a = tf.keras.layers.Dropout(0.8)(layer3a)

    merge = concatenate([drop1a, drop2a,    drop3a])
    outputs = tf.keras.layers.Dense(units= 1, activation='sigmoid')(merge)
    model_2 = tf.keras.Model(inputs=input, outputs=(outputs))




    return model_2
```

```python
model_2 = make_model_2()

model_2.compile(optimizer=keras.optimizers.Adam(lr=1e-3),
                  loss=keras.losses.BinaryCrossentropy(),
                  metrics=METRICS)


model_2.summary()
```

```python
tf.keras.utils.plot_model(model_2,  show_shapes=True)
```

In the module, the input from one layer is passed into three separate layers, which are then concatenated back into one output layer.

When two or more nodes are concatenated, they are put end-to-end into one large vector, which is passed into the next node. The concatenate layer takes in a list of nodes to concatenate, and the results of this layer are then passed into another outputs node.

```python
history_2 = model_2.fit(
    X_train_bal,
    y_train_bal,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    validation_data=(X_val, y_val), 
    #callbacks=[early_stopping],
    verbose=1)
```

```python
plt.rcParams["figure.figsize"] = (10,10)
plot_metrics(history_2)
```

```python
test_predictions_2 = model_2.predict(X_test, batch_size=BATCH_SIZE)
```

## Support Vector Machine

```python
from sklearn.svm import SVC


def SVMmodel(X, y):
    estimator_SVM = SVC(probability=True, random_state=1)

    parameters_SVM = {
        "C": [0.5, 1, 3, 5, 11, 20],
        "gamma": (0.1, 10.0, 1.0, 3),
        "kernel": ("linear", "rbf"),
    }

    # with GridSearch
    grid_search_SVM = GridSearchCV(
        estimator=estimator_SVM,
        param_grid=parameters_SVM,
        scoring="accuracy",
        n_jobs=-1,
        return_train_score=True,
        cv=5,
        verbose=3,
    )
    # fit the SVM model

    SVMmodel = grid_search_SVM.fit(X, y)

    # predicting on training and test data-set
    Y_pred = SVMmodel.predict(X)
    y_SVM_prob = grid_search_SVM.fit(X_train_bal, y_train_bal).predict_proba(X_test)[:,1] 

    # Assess the performance
    print("SVM : ")
    print(grid_search_SVM.best_params_)
    print("Averange Accuracy Scores - SVM:", metrics.accuracy_score(y, Y_pred))
    print("Confusion matrix - SVM: \n", confusion_matrix(y, Y_pred))
    return SVMmodel, y_SVM_prob


SVM_Model, y_SVM_prob = SVMmodel(    X_train_bal,    y_train_bal)
```

```python
y_pred_svm = SVM_Model.predict(X_test)
print('SVM: ')
print("Averange Accuracy Scores - SVM:", metrics.accuracy_score(y_test , y_pred_svm))
print('Precision score - SVM: ', precision_score(y_test, y_pred_svm))
print('Recall score score - SVM : ', recall_score(y_test, y_pred_svm))
print('F1 score - SVM: ',f1_score(y_test, y_pred_svm))
print('Confusion matrix - SVM: \n', confusion_matrix(y_test, y_pred_svm))
```
